```python
import tkinter as tk
from tkinter import messagebox
import random
import time

class QuizApp:
    def __init__(self, master):
        self.master = master
        master.title("DevOps/SRE Quiz")
        self.score = 0
        self.current_question_index = 0
        self.questions = self.load_questions()
        self.feedback = []

        self.start_time = time.time()
        self.total_time = 120  # 2 minutes

        # GUI Elements
        self.timer_label = tk.Label(master, text="", fg="red", font=("Arial", 12, "bold"))
        self.timer_label.pack(pady=(5, 0))

        self.question_label = tk.Label(master, text="", wraplength=600, justify="left")
        self.question_label.pack(pady=10)

        self.var = tk.IntVar(value=-1)
        self.radio_buttons = [tk.Radiobutton(master, variable=self.var, value=i, anchor='w') for i in range(4)]
        for rb in self.radio_buttons:
            rb.pack(fill="x", padx=20)

        self.next_button = tk.Button(master, text="Next", command=self.next_question)
        self.next_button.pack(pady=10)

        self.display_question()
        self.update_timer()

    def load_questions(self):
        return [
            ("What is CI/CD in DevOps?", 
             ["Continuous Integration/Continuous Deployment", "Configuration Inspection/Control Delivery", 
              "Continuous Implementation/Cluster Distribution", "Certified Integration/Certified Delivery"], 
             "Continuous Integration/Continuous Deployment"),
            
            ("What is a StatefulSet in Kubernetes used for?", 
             ["Managing stateful applications", "Load balancing services", "Monitoring node health", "Running cron jobs"], 
             "Managing stateful applications"),
            
            ("What does 'least privilege' mean in security?", 
             ["Minimal access required to perform a task", "Admin rights to all systems", "Frequent password changes", "Open access"], 
             "Minimal access required to perform a task"),
            
            ("What is the use of Helm in Kubernetes?", 
             ["Package manager for Kubernetes", "Load balancer", "Logging tool", "Cluster autoscaler"], 
             "Package manager for Kubernetes"),
            
            ("What is 'immutable infrastructure'?", 
             ["Servers replaced instead of modified", "Read-only containers", "Encrypted hardware", "Code that never changes"], 
             "Servers replaced instead of modified"),
            
            ("What does a service mesh primarily handle?", 
             ["Secure communication between services", "Static routing", "User authentication", "Data compression"], 
             "Secure communication between services"),

            ("What is the function of etcd in a Kubernetes cluster?", 
             ["Stores configuration and state data", "Routes network traffic", "Monitors logs", "Schedules pods"], 
             "Stores configuration and state data"),
            
            ("Which component schedules pods onto nodes in Kubernetes?", 
             ["Kube-scheduler", "Kubelet", "Controller Manager", "Etcd"], 
             "Kube-scheduler"),
            
            ("What is a liveness probe in Kubernetes?", 
             ["Checks if an app should be restarted", "Checks if app is ready for traffic", "Measures CPU usage", "Scales the deployment"], 
             "Checks if an app should be restarted"),
            
            ("What is a readiness probe?", 
             ["Checks if the app is ready to receive traffic", "Checks if app is dead", "Loads environment variables", "Creates volumes"], 
             "Checks if the app is ready to receive traffic"),
            
            ("What is the primary purpose of Infrastructure as Code (IaC)?", 
             ["Automate and version infrastructure", "Generate reports", "Backup databases", "Log system calls"], 
             "Automate and version infrastructure"),
            
            ("Which tool is commonly used for Infrastructure as Code?", 
             ["Terraform", "Kibana", "Docker", "Prometheus"], 
             "Terraform"),

            ("Which tool is used to collect, parse, and visualize logs?", 
             ["ELK Stack", "Helm", "etcd", "HPA"], 
             "ELK Stack")
        ]

    def display_question(self):
        q_text, options, correct = self.questions[self.current_question_index]
        self.correct_answer = correct
        self.shuffled_options = options[:]
        random.shuffle(self.shuffled_options)

        # Store correct answer's index in the shuffled list
        self.correct_index = self.shuffled_options.index(correct)

        self.question_label.config(text=f"Q{self.current_question_index + 1}: {q_text}")
        for i, opt in enumerate(self.shuffled_options):
            self.radio_buttons[i].config(text=opt)
        self.var.set(-1)

    def next_question(self):
        selected = self.var.get()
        if selected == -1:
            messagebox.showwarning("No selection", "Please select an answer or pass.")
            return

        # Compare selected index with stored correct index
        if selected == self.correct_index:
            self.score += 1

        selected_text = self.shuffled_options[selected]
        correct_text = self.shuffled_options[self.correct_index]
        q_text = self.questions[self.current_question_index][0]

        self.feedback.append(
            f"Q: {q_text}\nYour answer: {selected_text}\nCorrect answer: {correct_text}\n"
        )

        if self.current_question_index < len(self.questions) - 1:
            self.current_question_index += 1
            self.display_question()
        else:
            self.end_quiz()

    def update_timer(self):
        elapsed = int(time.time() - self.start_time)
        remaining = self.total_time - elapsed
        if remaining >= 0:
            self.timer_label.config(text=f"Time Left: {remaining} seconds")
            self.master.after(1000, self.update_timer)
        else:
            self.end_quiz()

    def end_quiz(self):
        self.timer_label.config(text="Time's up!")
        feedback_text = "\n".join(self.feedback)
        messagebox.showinfo("Quiz Complete", f"Your Score: {self.score}/{len(self.questions)}\n\n{feedback_text}")
        self.master.quit()
        self.master.destroy()

# Run the app
root = tk.Tk()
root.geometry("700x450")
app = QuizApp(root)
root.mainloop()
```
